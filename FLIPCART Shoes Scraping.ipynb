The goal of this data scraping project is to collect comprehensive information about shoes available on Flipkart. 
By employing web scraping techniques, we will automatically extract key data points such as shoe descriptions,
shoe names, pricing, and price_off on these shoes.

```python
#importing all the necessary libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup

#Creating empty lists to store scraped data.

product_names=[]
product_prices=[]
product_description=[]
product_priceoff=[]

```

```python
#for i in range(2,10):
 
# url of the website to get access to 

url="https://www.flipkart.com/search?q=shoes+under+500&sid=osp%2Ccil%2Ce1f&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_4_10_sc_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_4_10_sc_na_na&as-pos=4&as-type=RECENT&suggestionId=shoes+under+500%7CCasual+Shoes&requestId=cfa2ae55-ade1-4785-9bde-2ccdf4083dbe&as-searchtext=shoes+under+500&page=1"
r=requests.get(url)

#Creating a soup variable to get data
soup=BeautifulSoup(r.text, "lxml")

#box=soup.find("div", class_="_1YokD2 _3Mn1Gg")

#now creating a box varible to store the data extracted from a specific div
box=soup.find("div", class_="_1YokD2 _3Mn1Gg")

# box.find_all to get the names and then save into the names varible.
names=box.find_all("div", class_="_2WkVRV")
#print(names)

#for loop to iterate through the names and then append the name to the list product_names

for i in names:
    name=i.text
    product_names.append(name)

print(product_names)
print(len(product_names))


#Similar to the names here we will get the prices and append it to the list named product_prices 

prices=box.find_all("div", class_="_30jeq3")
#print(prices)

for i in prices:
    price=i.text
    product_prices.append(price)

print(product_prices)
print(len(product_prices))


#Similar to the above situations

description=box.find_all("a", class_="IRpwTa")
#print(description)

for i in description:
    desc=i.text
    product_description.append(desc)

print(product_description)
print(len(product_description))

#same as the name, prices and description

price_off=box.find_all("div", class_="_3Ay6Sb")
#print(price_off)
for i in price_off:
    po=i.text
    product_priceoff.append(po)

print(product_priceoff)
print(len(product_priceoff))


#creating a dataframe to store values in a nice manner

df=pd.DataFrame({"Product Name" : product_names, "Product Price": product_prices , "Product Description": product_description, "price off on products": product_priceoff})
print(df)

#Converting the dartframe into a csv file

df.to_csv("Shoes on Flipcart.csv")
        

#new_page=soup.find("a", class_="_1LKTO3").get("href")
#newpage_link="https://www.flipkart.com"+ new_page 
#print(newpage_link)
```
